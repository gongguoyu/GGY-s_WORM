```python
import matplotlib.pyplot as plt     #数学绘图库
%matplotlib inline
import jieba               #分词库
from wordcloud import WordCloud   #词云库
```

```python
names=''
contents=''

f = open("./QQMessage.txt")
lines = f.readlines()
for line3,i in zip(lines,range(len(lines))):
    if ') ' in line3:
        print(i,'N',line3[:-1])
        names+=line3[:-1].split('(')[0]+'/'
    elif '> ' in line3:
        print(i,'N',line3[:-1])
        names+=line3[:-1].split('<')[0]+'/'
    else:
        #print(i,':',line3[:-1])
        contents+=line3[:-1]
print('Done!')
```

```python
print(names)

print('------------')


contents= contents.replace('双击','')
contents= contents.replace('查看','')
contents= contents.replace('原图','')
contents= contents.replace('加入','')
contents= contents.replace('本群','')
result_content= "/".join(jieba.cut(contents))
print(result_content)
```

```python
wc = WordCloud(font_path="msyhbd.ttf",
               background_color='white',width=400,height=300,max_font_size=50,min_font_size=5,
               max_words=100)#,min_font_size=10)#,mode='RGBA',colormap='pink')
wc.generate(names)
plt.figure(figsize =(8,6)) #指定所绘图名称
plt.imshow(wc)       # 以图片的形式显示词云
plt.axis("off")      #关闭图像坐标系
plt.show()
```

```python
wc = WordCloud(font_path="msyhbd.ttf",
               background_color='white',width=400,height=300,max_font_size=50,min_font_size=8,
               max_words=200)#,min_font_size=10)#,mode='RGBA',colormap='pink')
wc.generate(result_content)
plt.figure(figsize =(8,6)) #指定所绘图名称
plt.imshow(wc)       # 以图片的形式显示词云
plt.axis("off")      #关闭图像坐标系
plt.show()
```

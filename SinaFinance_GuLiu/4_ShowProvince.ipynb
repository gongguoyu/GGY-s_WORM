```python
def get_pro(area):
    prolist = ['åŒ—äº¬', 'å¤©æ´¥', 'ä¸Šæµ·', 'é‡åº†', 'æ²³åŒ—', 'å±±è¥¿', 'è¾½å®', 'å‰æ—', 'æ±Ÿè‹', 'æµ™æ±Ÿ', 'å®‰å¾½', 'ç¦å»º', 'æ±Ÿè¥¿', 'å±±ä¸œ', 'æ²³å—', 'æ¹–åŒ—', 'æ¹–å—', 'å¹¿ä¸œ', 'æµ·å—', 'å››å·', 'è´µå·', 'äº‘å—', 'é™•è¥¿', 'ç”˜è‚ƒ', 'é’æµ·', 'å°æ¹¾', 'å¹¿è¥¿', 'è¥¿è—', 'å®å¤', 'æ–°ç–†', 'é¦™æ¸¯', 'æ¾³é—¨', 'å†…è’™å¤', 'é»‘é¾™æ±Ÿ']
    for pro in prolist:
        if pro in area:
            return pro
    return "æµ·å¤–"
df['pro'] = df.area.apply(get_pro)
df[['area','pro']]
```

```python
pro_count = df.groupby('pro')['pro'].count().sort_values(ascending=False)
pro_count
```

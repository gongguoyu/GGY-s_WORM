```python
root_name = 'Sina_Finance_Comments_85C_20180823'
print(root_name)
```

```python
import pandas as pd
df = pd.read_csv(root_name+'_Cleaned.csv',encoding='utf-8')
df.head()
```

```python
message = ''

for row in df.content:
    message+=row

print(message)
```

```python
import matplotlib.pyplot as plt     #数学绘图库
%matplotlib inline
import jieba               #分词库
from wordcloud import WordCloud   #词云库
```

```python
cut_text= jieba.cut(message)

result= "/".join(cut_text)#必须给个符号分隔开分词结果来形成字符串,否则不能绘制词云
print(result)
```

```python
wc = WordCloud(font_path="./Test_CloudGraph/msyh.ttf",
               background_color='white',width=400,height=300,max_font_size=40,min_font_size=8,
               max_words=400)#,min_font_size=10)#,mode='RGBA',colormap='pink')
wc.generate(result)
```

```python
# 4、显示图片
plt.figure(figsize =(8,6)) #指定所绘图名称
plt.imshow(wc)       # 以图片的形式显示词云
plt.axis("off")      #关闭图像坐标系
plt.show()
```
